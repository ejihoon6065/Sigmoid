```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# x : [공부한 시간, 출석일수]

xdata = [[1,3], [2,2], [3,1], [4,6], [5,5], [6,4]]
ydata = [[0], [0], [0], [1], [1], [1]] # 0이면 불합격, 1이면 합격
```

```python
x = tf.placeholder( tf.float32, shape=[None,2])
y = tf.constant( ydata, tf.float32)
```

```python
w = tf.Variable( tf.random_uniform([2,1]) )
b = tf.Variable( tf.random_uniform([1]) )
```

```python
# z는 연속된 데이터
z = tf.matmul(x,w) + b

# sigmoid 통과하면 0과 1로 도출됨
hx = tf.nn.sigmoid( z ) # 0~1사이의 값
```

```python
# tensor에서 로그 함수 제공
# 오타 유의

cost_i = y*(-tf.log(hx)) + (1-y)*(-tf.log(1-hx))
cost = tf.reduce_mean( cost_i )
```

```python
optimizer = tf.train.AdamOptimizer( 0.1 ) 
train = optimizer.minimize( cost )

sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)
```

```python
for i in range(1000) :
    sess.run( train, {x:xdata})
    print( i, sess.run(cost, {x:xdata}))
```

```python
sess.run(b)
```

```python
result = sess.run(hx, {x:[[3,5]]})
```

```python
# cast함수 : true인 경우 1, false인 경우 0으로 출력

t = tf.cast( result > 0.5, dtype=tf.int32 )
sess.run(t)
```

```python
# 전체 데이터 예측; x:6x2 *(matmul)* w:2x1 -> 6x1 행렬

h = sess.run(hx, {x:xdata})
h
```

```python
(h > 0.5) == ydata
# True는 1, False는 0이므로  
```

```python
((h > 0.5) == ydata).mean() # 정확도를 의미 => 1.0 = 100퍼센트
```

```python
# y가 float로 되어 있음

predict = tf.cast(h>0.5, dtype = tf.float32)

# sess.run(predict)
e = tf.equal(predict, y) # predict == y랑 같다

m = tf.reduce_mean( tf.cast(e, dtype=tf.float32))

sess.run(m) # 1.0 -> 예측값 100퍼센트
```
